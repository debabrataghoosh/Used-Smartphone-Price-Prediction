```python
# Cell 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style for plots
sns.set_style("whitegrid")

print("Libraries imported successfully!")
```

```python
# Cell 2: Load your dataset
# IMPORTANT: Replace 'your_file_name.csv' with the exact name of your uploaded file.
file_path = '/content/Data/smartphones_data.csv.csv'

try:
    df = pd.read_csv(file_path)
    print(f"✅ Successfully loaded {file_path}")
except FileNotFoundError:
    print(f"❌ Error: The file '{file_path}' was not found.")
    print("Please make sure you have uploaded the file and the name is spelled correctly.")
```

```python
# Cell 3: Check the shape and view the first few rows
print(f"The dataset has {df.shape[0]} rows and {df.shape[1]} columns.\n")
print("--- First 5 Rows ---")
display(df.head())
```

```python
# Cell 4: Get detailed info about each column
print("\n--- Dataset Info ---")
df.info()

# ==> From the output above, note your actual column names. You'll need them below!
```

```python
# Cell 5: Check for missing values
print("--- Missing Values Before Cleaning ---")
print(df.isnull().sum())
```

```python
# A dictionary specifying the default value for each column
fill_values = {
    'has_fingerprints': 0,  # 0 represents "no"
    'has_nfc': 0,           # 0 represents "no"
    'has_5g': 0,            # 0 represents "no"
    'num_core': 4,
    'refresh_rate(hz)': 60
}

# Fill the missing values in the DataFrame
df.fillna(value=fill_values, inplace=True)

# Optional: Print the missing value counts again to confirm they are filled
print("--- Missing Values After Cleaning ---")
print(df.isnull().sum())
```

```python
# Check how many duplicate rows exist
print(f"Number of duplicate rows found: {df.duplicated().sum()}")

# Remove the duplicates
df.drop_duplicates(inplace=True)

print(f"Dataset shape after dropping duplicates: {df.shape}")
```

```python
# Analyze the 'processor_brand' column
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='processor_brand', order=df['processor_brand'].value_counts().index, palette='viridis')
plt.title('Processor Brand Distribution', fontsize=16)
plt.xlabel('Count')
plt.ylabel('Processor Brand')
plt.show()
```

```python
# Analyze the 'Price' column
plt.figure(figsize=(12, 5))

# Histogram to see the price distribution
plt.subplot(1, 2, 1)
sns.histplot(df['Price'], kde=True, bins=25, color='skyblue')
plt.title('Smartphone Price Distribution', fontsize=14)
plt.xlabel('Price')

# Box plot to spot outliers
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Price'], color='lightcoral')
plt.title('Price Outlier Check', fontsize=14)
plt.xlabel('Price')

plt.tight_layout()
plt.show()
```

```python
# Select only numeric columns for the correlation matrix
numeric_df = df.select_dtypes(include=np.number)
correlation_matrix = numeric_df.corr()

# Visualize the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Smartphone Features', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='RAM', y='Price', hue='processor_brand', alpha=0.7)
plt.title('RAM vs. Price', fontsize=14)
plt.xlabel('RAM (GB)')
plt.ylabel('Price')
plt.show()
```
